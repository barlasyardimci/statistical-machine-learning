```python
import numpy
import scipy.io
import math
import geneNewData

def main():
    myID='9330'
    geneNewData.geneData(myID)
    Numpyfile0 = scipy.io.loadmat('digit0_stu_train'+myID+'.mat')
    Numpyfile1 = scipy.io.loadmat('digit1_stu_train'+myID+'.mat')
    Numpyfile2 = scipy.io.loadmat('digit0_testset'+'.mat')
    Numpyfile3 = scipy.io.loadmat('digit1_testset'+'.mat')
    train0 = Numpyfile0.get('target_img')
    train1 = Numpyfile1.get('target_img')
    test0 = Numpyfile2.get('target_img')
    test1 = Numpyfile3.get('target_img')
    print([len(train0),len(train1),len(test0),len(test1)])
    print('Your trainset and testset are generated successfully!')
    
    #Task 1
    
    #calculations for feature 1
    
    #average of pixels for 0s
    averagePixelArray0 = train0.mean(axis = (1,2)).reshape(5000,1)
    print(averagePixelArray0.shape)
    #average of pixels for 1s
    averagePixelArray1 = train1.mean(axis = (1,2)).reshape(5000,1)
    print(averagePixelArray1.shape)
    
    #calculations for feature 2
    
    #standard deviation for 0s and 1s
    stdArray0 = train0.std(axis = (1,2)).reshape(5000,1)
    stdArray1 = train1.std(axis = (1,2)).reshape(5000,1)
    print(stdArray0.shape)
    print(stdArray1.shape)
    
    #Task 2
    
    #No1  Mean of feature1 for digit0
    meanF1D0 = averagePixelArray0.mean()
    print(meanF1D0)
    #No2 Variance of feature1 for digit0
    varF1D0 = averagePixelArray0.var()
    print(varF1D0)
    #No3 Mean of feature2 for digit0
    meanF2D0 = stdArray0.mean()
    print(meanF2D0)
    #No4 Variance of feature2 for digit0
    varF2D0 = stdArray0.var()
    print(varF2D0)
    #No5 Mean of feature1 for digit1
    meanF1D1 = averagePixelArray1.mean()
    print(meanF1D1)
    #No6 Variance of feature1 for digit1
    varF1D1 = averagePixelArray1.var()
    print(varF1D1)
    #No7 Mean of feature2 for digit1
    meanF2D1 = stdArray1.mean()
    print(meanF2D1)
    #No8 Variance of feature2 for digit1
    varF2D1 = stdArray1.var()
    print(varF2D1)
    
    #Task 3
    
    averagePixelTestArray0 = test0.mean(axis = (1,2)).reshape(test0.shape[0],1)
    print(averagePixelTestArray0.shape)
    
    averagePixelTestArray1 = test1.mean(axis = (1,2)).reshape(test1.shape[0],1)
    print(averagePixelTestArray1.shape)
    
    stdTestArray0 = test0.std(axis = (1,2)).reshape(test0.shape[0],1)
    stdTestArray1 = test1.std(axis = (1,2)).reshape(test1.shape[0],1)
    print(stdTestArray0.shape)
    print(stdTestArray1.shape)
    
    #In order to calculate vals prior x feature1-likelihood x feature2-likelihood
    digitZeroScoresTestZero = numpy.log(0.5 * 1/numpy.sqrt(2*math.pi*varF1D0) * numpy.exp(-(averagePixelTestArray0-meanF1D0)**2/(2*varF1D0)) * 1/numpy.sqrt(2*math.pi*varF2D0) * numpy.exp(-(stdTestArray0-meanF2D0)**2/(2*varF2D0)).reshape(test0.shape[0],1))
    digitOneScoresTestZero = numpy.log(0.5 * 1/numpy.sqrt(2*math.pi*varF1D1) * numpy.exp(-(averagePixelTestArray0-meanF1D1)**2/(2*varF1D1)) * 1/numpy.sqrt(2*math.pi*varF2D1) * numpy.exp(-(stdTestArray0-meanF2D1)**2/(2*varF2D1)).reshape(test0.shape[0],1))
    print(digitZeroScoresTestZero.shape)
    print(digitOneScoresTestZero.shape)
    
    digitZeroScoresTestOne = numpy.log(0.5 * 1/numpy.sqrt(2*math.pi*varF1D0) * numpy.exp(-(averagePixelTestArray1-meanF1D0)**2/(2*varF1D0)) * 1/numpy.sqrt(2*math.pi*varF2D0) * numpy.exp(-(stdTestArray1-meanF2D0)**2/(2*varF2D0)).reshape(test1.shape[0],1))
    digitOneScoresTestOne = numpy.log(0.5 * 1/numpy.sqrt(2*math.pi*varF1D1) * numpy.exp(-(averagePixelTestArray1-meanF1D1)**2/(2*varF1D1)) * 1/numpy.sqrt(2*math.pi*varF2D1) * numpy.exp(-(stdTestArray1-meanF2D1)**2/(2*varF2D1)).reshape(test1.shape[0],1))
    print(digitZeroScoresTestOne.shape)
    print(digitOneScoresTestOne.shape)
    
    labels_0 = numpy.empty([test0.shape[0],1])
    labels_1 = numpy.empty([test1.shape[0],1])
    
    #predict for test0
    for  row in range(len(test0)):
        if digitZeroScoresTestZero[row][0] < digitOneScoresTestZero[row][0]:
            labels_0[row][0] = 1
        else:
            labels_0[row][0] = 0
    
    #predict for test1
    for  row in range(len(test1)):
        if digitZeroScoresTestOne[row][0] < digitOneScoresTestOne[row][0]:
            labels_1[row][0] = 1
        else:
            labels_1[row][0] = 0
    
    #Calculate number of classified digits
    prediction0_test0 = (labels_0==0).sum()
    prediction1_test0 = len(labels_0) - prediction0_test0
    print(prediction0_test0)
    print(prediction1_test0)
    
    #Calculate number of classified digits
    prediction0_test1 = (labels_1==0).sum()
    prediction1_test1 = len(labels_1) - prediction0_test1
    print(prediction0_test1)
    print(prediction1_test1)
    
    #Task 4 accuracy calculation
    test0_accuracy = prediction0_test0 / (prediction0_test0 + prediction1_test0)
    test1_accuracy = prediction1_test1 / (prediction0_test1 + prediction1_test1)
    print(test0_accuracy)
    print(test1_accuracy)
    
if __name__ == '__main__':
    main()
```
